```
madrid_df = spark.read.format("csv") \
            .option("header","true") \
            .option("inferSchema","true") \
            .load("/FileStore/tables/madrid_tweet_data-1.csv")
            
madrid_df.show()
```

```
barca_df = spark.read.format("csv") \
            .option("header","true") \
            .option("inferSchema", "true") \
            .load("/FileStore/tables/barca_tweet_data-1.csv")
            
barca_df.count()
```

```
teams_df_1 = madrid_df.union(barca_df)

teams_df_1.count()
```

```
from pyspark.sql.functions import *
```

```
teams_df = teams_df_1.withColumn("tweet_date", date_format(col("created_at"), "yyyy-MM-dd"))

teams_df.show()
```

```


teams_df = teams_df.select('user', col('id').alias('tweet_id'), 'lang', 'source',
                  col('favorite_count').alias('likes'), col('retweet_count').alias('retweets'), 'tweet_date')

teams_df.show()
```
